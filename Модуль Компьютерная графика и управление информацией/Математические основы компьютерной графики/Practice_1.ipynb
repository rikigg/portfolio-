```python
import numpy as np
import pprint
import pygame
import sys

# ввод и преобразование координат
x = int(input('Введите X: '))
y = int(input('Введите Y: '))

xy = np.array([[x, y]])

T = np.array([
    [1,3],
    [4,1]]
    )

x1y1 = xy @ T

x1 = x1y1[0, 0]
y1 = x1y1[0, 1]

# печать в консоль
pp = pprint.PrettyPrinter(width=20)
print(f'Исходные координаты: X = {x}, Y = {y}')
pp.pprint(T)
print(f'Полученные координаты: X1 = {x1}, Y1 = {y1}')

# Прорисовка
pygame.init()

screen = pygame.display.set_mode((1200, 700))
screen.fill((255,255,255))

my_font = pygame.font.SysFont('Comic Sans MS', 25)

# исходная точка
pygame.draw.circle(screen, (60, 115, 20), (x, y), 8, width=0)
cord1 = my_font.render(str(f'Введённая: ({x},{y})'), True, (60, 115, 20))
screen.blit(cord1, (x+10, y-40))

# полученная точка
pygame.draw.circle(screen, (40, 20, 140), (x1, y1), 8, width=0)
cord2 = my_font.render(str(f'Полученная: ({x1},{y1})'), True, (40, 20, 140))
screen.blit(cord2, (x1+10, y1-40))

FPS = 30
clock = pygame.time.Clock()
while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
   
    clock.tick(FPS)
    pygame.display.update()
```

```python
import numpy as np
import pprint
import pygame
import sys

# ввод и преобразование координат
x1 = int(input('Введите X1: '))
y1 = int(input('Введите Y1: '))
x2 = int(input('Введите X2: '))
y2 = int(input('Введите Y2: '))

line1 = np.array([[x1, y1], [x2, y2]])

T = np.array([
    [1,3],
    [4,1]]
    )

line2 = line1 @ T

# печать в консоль
pp = pprint.PrettyPrinter(width=20)
print('Исходная линия:')
pp.pprint(line1)
print('Матрица преобразования:')
pp.pprint(T)
print('Полученная линия:')
pp.pprint(line2)

# Прорисовка
pygame.init()

screen = pygame.display.set_mode((1200, 700))
screen.fill((255,255,255))

my_font = pygame.font.SysFont('Comic Sans MS', 20)

# исходная линия
pygame.draw.line(screen, (60, 115, 20), (line1[0, 0], line1[0, 1]), (line1[1, 0], line1[1, 1]), 5)
line1_coord1 = my_font.render(str(f'({line1[0, 0]},{line1[0, 1]})'), True, (60, 115, 20))
screen.blit(line1_coord1, (line1[0, 0], line1[0, 1]))
line1_coord2 = my_font.render(str(f'({line1[1, 0]},{line1[1, 1]})'), True, (60, 115, 20))
screen.blit(line1_coord2, (line1[1, 0], line1[1, 1]))

# полученная линия
pygame.draw.line(screen, (40, 20, 140), (line2[0, 0], line2[0, 1]), (line2[1, 0], line2[1, 1]), 5)
line2_coord1 = my_font.render(str(f'({line2[0, 0]},{line2[0, 1]})'), True, (40, 20, 140))
screen.blit(line2_coord1, (line2[0, 0], line2[0, 1]))
line2_coord2 = my_font.render(str(f'({line2[1, 0]},{line2[1, 1]})'), True, (40, 20, 140))
screen.blit(line2_coord2, (line2[1, 0], line2[1, 1]))

FPS = 30
clock = pygame.time.Clock()
while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
   
    clock.tick(FPS)
    pygame.display.update()
```
