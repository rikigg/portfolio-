python task_1_1.py

```python
import numpy as np
import pprint
import pygame
import sys

L = np.array([[0, 100], [200, 300]])

mid_point = np.array([[(L[0, 0]+L[1, 0])/2, (L[0, 1]+L[1, 1])/2]])

T = np.array([
    [1,2],
    [3,1]]
    )

L_conv = L @ T
mid_point_conv = mid_point @ T

# печать в консоль
pp = pprint.PrettyPrinter(width=20)
print('Исходный отрезок:')
pp.pprint(L)
print('Середина:')
pp.pprint(mid_point)
print('Матрица преобразования:')
pp.pprint(T)
print('Полученный отрезок:')
pp.pprint(L_conv)
print('Середина полученного отрезка:')
pp.pprint(mid_point_conv)

# Прорисовка

pygame.init()

screen = pygame.display.set_mode((1280, 720))
screen.fill((255,255,255))
my_font = pygame.font.SysFont('Comic Sans MS', 20)

# исходный отрезок
pygame.draw.line(screen, (60, 115, 20), (L[0, 0], L[0, 1]), (L[1, 0], L[1, 1]), 5)
pygame.draw.circle(screen, (60, 115, 20), (mid_point[0, 0], mid_point[0, 1]), 7)

# полученный отрезок
pygame.draw.line(screen, (40, 20, 140), (L_conv[0, 0], L_conv[0, 1]), (L_conv[1, 0], L_conv[1, 1]), 5)
pygame.draw.circle(screen, (40, 20, 140), (mid_point_conv[0, 0], mid_point_conv[0, 1]), 7)

FPS = 30
clock = pygame.time.Clock()
while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
   
    clock.tick(FPS)
    pygame.display.update()
```

```python
import numpy as np
import pprint
import pygame
import sys

L = np.array([[50, 100], [250, 200], [50, 200], [250, 300]])

m1_AB = (L[1, 1] - L[0, 1]) / (L[1, 0] - L[0, 0])
m1_EF = (L[3, 1] - L[2, 1]) / (L[3, 0] - L[2, 0])

T = np.array([
    [1,2],
    [3,1]]
    )

L_conv = L @ T

m1_A1B1 = (L_conv[1, 1] - L_conv[0, 1]) / (L_conv[1, 0] - L_conv[0, 0])
m1_E1F1 = (L_conv[3, 1] - L_conv[2, 1]) / (L_conv[3, 0] - L_conv[2, 0])

# печать в консоль
pp = pprint.PrettyPrinter(width=20)
print(f'Исходные отрезки: AB = [[{L[0, 0], L[0, 1]}], [{L[1, 0], L[1, 1]}]], EF = [[{L[2, 0], L[2, 1]}], [{L[3, 0], L[3, 1]}]]')
print(f'Наклон AB = {m1_AB}, Наклон EF = {m1_EF}')

print('Матрица преобразования:')
pp.pprint(T)

print(f'Полученные отрезки: A1B1 = [[{L_conv[0, 0], L_conv[0, 1]}], [{L_conv[1, 0], L_conv[1, 1]}]], E1F1 = [[{L_conv[2, 0], L_conv[2, 1]}], [{L_conv[3, 0], L_conv[3, 1]}]]')
print(f'Наклон A1B1 = {m1_A1B1}, Наклон E1F1 = {m1_E1F1}')

# Прорисовка

pygame.init()

screen = pygame.display.set_mode((1280, 720))
screen.fill((255,255,255))
my_font = pygame.font.SysFont('Comic Sans MS', 20)

# исходные отрезки
pygame.draw.line(screen, (60, 115, 20), (L[0, 0], L[0, 1]), (L[1, 0], L[1, 1]), 5)
pygame.draw.line(screen, (60, 115, 20), (L[2, 0], L[2, 1]), (L[3, 0], L[3, 1]), 5)

# полученные отрезки
pygame.draw.line(screen, (40, 20, 140), (L_conv[0, 0], L_conv[0, 1]), (L_conv[1, 0], L_conv[1, 1]), 5)
pygame.draw.line(screen, (40, 20, 140), (L_conv[2, 0], L_conv[2, 1]), (L_conv[3, 0], L_conv[3, 1]), 5)


FPS = 30
clock = pygame.time.Clock()
while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
   
    clock.tick(FPS)
    pygame.display.update()
```

```python
import numpy as np
import pprint
import pygame
import sys

L = np.array([[3, -1], [4, 1], [2, 1]])
L = L*100

T = np.array([
    [0, 1],
    [-1, 0]]
    )

L_conv = L @ T

# печать в консоль
pp = pprint.PrettyPrinter(width=20)
print(f'Исходный треугольник (green):[[{round(L[0, 0],2), round(L[0, 1], 2)}], [{round(L[1, 0], 2), round(L[1, 1], 2)}], [{round(L[2, 0], 2), round(L[2, 1], 2)}]]')

print('Матрица преобразования (поворот):')
pp.pprint(T)

print(f'Полученный треугольник (blue): [[{round(L_conv[0, 0],2), round(L_conv[0, 1], 2)}], [{round(L_conv[1, 0], 2), round(L_conv[1, 1], 2)}], [{round(L_conv[2, 0], 2), round(L_conv[2, 1], 2)}]]')

# Прорисовка

pygame.init()

screen = pygame.display.set_mode((1280, 720))
screen.fill((255,255,255))
my_font = pygame.font.SysFont('Comic Sans MS', 20)

# точка начала координат
pygame.draw.circle(screen, (0, 0, 0), (0, 0), 15)

# исходный треугольник
pygame.draw.polygon(screen, (60, 115, 20), [(L[0, 0] + 200, L[0, 1] + 200), (L[1, 0] + 200, L[1, 1] + 200), (L[2, 0] + 200, L[2, 1] + 200)], 5)

# полученный треугольник
pygame.draw.polygon(screen, (40, 20, 140), [(L_conv[0, 0] + 200, L_conv[0, 1] + 200), (L_conv[1, 0] + 200, L_conv[1, 1] + 200), (L_conv[2, 0] + 200, L_conv[2, 1] + 200)], 5)

FPS = 30
clock = pygame.time.Clock()
while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
   
    clock.tick(FPS)
    pygame.display.update()
```
