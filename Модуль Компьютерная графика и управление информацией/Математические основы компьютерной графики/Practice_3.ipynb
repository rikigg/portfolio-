```python
import numpy as np
import pprint
import pygame
import sys
import math

#------------------------

pygame.init()

screen = pygame.display.set_mode((800, 800))
screen.fill((255,255,255))
my_font = pygame.font.SysFont('Comic Sans MS', 20)

#------------------------

margin = 400

X = np.array([[2, 2], [-2, 2], [-2, -2], [2, -2]]) * 100

print('Вершины квадрата: ')
pprint.pprint(X)

m = 0.9

M = np.array([
    [m, 0],
    [0, m]]
    )

alpha = -1 * math.pi / 32
V = np.array([
    [math.cos(alpha), -math.sin(alpha)],
    [math.sin(alpha), math.cos(alpha)]]
    )

red = 60
green = 115
blue = 20

pygame.draw.polygon(screen, (red, green, blue), X + margin, 2)

T = V @ M

n = 30
X_iter = X.T
for i in range(1, n+1):
    X_iter= T @ X_iter
    red += 5
    pygame.draw.polygon(screen, (red, green, blue), (X_iter + margin).T, 2)

#------------------------

FPS = 30
clock = pygame.time.Clock()
while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
   
    clock.tick(FPS)
    pygame.display.update()
```

```python
import numpy as np
import pygame
import sys
import math

#------------------------

pygame.init()

screen = pygame.display.set_mode((800, 800))
screen.fill((255,255,255))
my_font = pygame.font.SysFont('Comic Sans MS', 20)

#------------------------

margin = 300

red = 60
green = 115
blue = 20

for b in range(10, 100 , 15):

    red -= 6
    green -= 10
    blue += 20

    a = 100
    alpha = 0

    d_alpha = math.pi / 32

    coords = list()

    for i in range(70):
        r = b + 2 * a * math.cos(alpha)
        x = r * math.cos(alpha)
        y = r * math.sin(alpha)
        coords.append([x, y])
        alpha += d_alpha

    coords_np_array = np.array(coords)
    pygame.draw.lines(screen, (red, green, blue), False, coords_np_array + margin, 2)

#------------------------

FPS = 30
clock = pygame.time.Clock()
while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
   
    clock.tick(FPS)
    pygame.display.update()
```
